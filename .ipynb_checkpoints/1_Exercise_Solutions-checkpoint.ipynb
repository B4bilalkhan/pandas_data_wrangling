**Exercise 1.3: Indexing Series and Acessing Series Data**


---


Which lines of code will create the `Series`, `shopping_list`, which has two entries: '2' and '1' labeled by the index `['musubi', 'wax' ]`, and then print the value labeled by 'wax'?

The `Series` should have the following form:

|  | |
|:----------|-----------|
| `musubi` | 2 |
| `wax` | 1  |


A:
```python
shopping_list = pd.Series({'musubi': 2, 'wax': 1}) 
shopping_list[0]
```

B:
```python
shopping_list = pd.Series( ['musubi', 'wax'], index=[2,1]) 
shopping_list[0]
```

C:
```python
shopping_list = pd.Series( [2,1], index=['musubi', 'wax']) 
shopping_list['wax']
```

D:
```python
shopping_list = pd.Series({2:1, 'index':['musubi', 'wax']}) 
shopping_list.wax
```

**Correct Answer**

C:
```python
shopping_list = pd.Series( [2,1], index=['musubi', 'wax']) 
shopping_list['wax']
```

**Explanation**

A: These lines of code will correctly build the `shopping_list Series`, but,  will print the value labeled by 'musubi' rather than 'wax'.

B: These lines of code will build the `Series` such that the index is the list `[2,1]` and the data entries are `['musubi', 'wax']`, but the problem statement clearly states that the index should be `['musubi', 'wax']` and the data entries should be `[2,1]`. 

C: This is the correct answer. The index of the `Series` is set using the `index` parameter of the `pandas Series()` function. The value labeled by `wax` is accessed using the square bracket notation and the label.

D: These lines of code will breate a `Series` with the following form:


|  | |
|:----------|-----------|
| `2` | 1 |
| `index` | \['musubi', 'wax'] |

and then since there is no 'wax' label in the `index` object of the `Series`, the line `shopping_list.wax` will cause an AttributeError: 'Series' object has no attribute 'wax'.

Which lines of code will access and print the zero'th row of `lunch_order_df`?

A:
```python
lunch_order_df.loc[['items', 'price'], 'Kalani']
```

B:
```python
lunch_order_df.iloc[0, :]
```

C:
```python
lunch_order_df.Kalani
```

D:
```python
lunch_order_df['Kalani']
```

**Correct Answer**

B:
```python
lunch_order_df.iloc[0, :]
```

**Explanation**

A: This line of code will cause a `KeyError` to be thrown with the message 'the label \[Kalani] is not in the \[columns]'. This is because the first positional argument in the square brackets of the `loc` attribute are for the index labels of the rows, and the second  positional argument is for the columns. The columns of `lunch_order_df` are `item` and `price` while the index labels are `Kalani` and `Tito`.

B: This is the correct answer. The problem states that the zero'th row of `lunch_order_df` should be accessed and printed. The `iloc` attribute allows accessing data by its integer location. Therefore, 0 is passed as the first positional argument of the `iloc` attribute and a colon,:, is passed as the second. Recall that the colon, :, is used to specify that we want to access all the columns.

C: This line of code causes a `AttributeError` to be thrown with the message, 'DataFrame' object has no attribute 'Kalani'. This is because the dot notation can only be used to access columns of a `DataFrame`. 

D: This line of code causes as `KeyError` to be thrown with the message 'Kalani'. This is because using brackets, \[], without the `iloc` or `loc` attributes can only be done to access columns. 

**Exercise 1.6: Adding Columns to DataFrames**

---


Which lines of code will add a column with the label: 'drink',  with all values set to 'water' to the 'lunch_order_df DataFrame' from Exercise 1.4 **in place** (the modification should be made to the original `DataFrame`)?

A:
```python
lunch_order_df.assign(drink = 'water')
```

B:
```python
lunch_order_df.loc[:, 'drink'] = 'water'
```

C:
```python
lunch_order_df.iloc[:, 2] = ['water', 'water']
```

D:
```python
lunch_order_df.assign(drink = ['water', 'water'])
```

**Correct Answer**

B:
```python
lunch_order_df.loc[:, 'drink'] = 'water'
```

**Explanation**

A, D: The `assign()` function will not modify the calling `DataFrame` in place, rather, it will make a copy of the `DataFrame` with the additional column. 

B: This is the correct answer. This line will create a new column labeled by 'drink' and assign the new data entries the value 'water' by broadcasting.

C: This answer will create a new column labeled by 2 and assign the new data entries the value 'water' using the list. This answer is incorrect since the problem statement clearly specifies that the new column should be labeled by 'drink'.

**Exercise 1.7: Adding Rows to DataFrames**

---


Which lines of code will make a **copy** of the 'lunch_order_df DataFrame' from Exercise 1.4,  and add a row with the label: 'Moana' and values 'Salad', 9.00, 'tea', to the columns 'item', 'price', and 'drink' respectively?

A:
```python
data = pd.Series({'item': 'Salad', 'price':9.00, 'drink': 'tea'}, name='Moana' )
lunch_order_df.iloc[2, :] = data
```

B:
```python
data = pd.Series({'item': 'Salad', 'price':9.00, 'drink': 'tea'})
lunch_order_df.append(data)
```

C:
```python
data = pd.Series({'item': 'Salad', 'price':9.00, 'drink': 'tea'})
lunch_order_df.loc['Moana', :] = data
```

D:
```python
data = pd.Series({'item': 'Salad', 'price':9.00, 'drink': 'tea'}, name='Moana')
lunch_order_df.append(data)
```

**Correct Answer**

D:
```python
data = pd.Series({'item': 'Salad', 'price':9.00, 'drink': 'tea'}, name='Moana')
lunch_order_df.append(data)
```

**Explanation**

A: This line of code causes an `IndexError` with the message 'single positional indexer is out-of-bounds'. This is because the `lunch_order_df.iloc[2, :] = data` command is attempting to access the row at integer location 2, which does not yet exist, hence the indexer, 2, is out-of-bounds. We cannot reference and assign new rows using the `iloc` attribute.

B: This line of code does not specify the new row's label which should be 'Moana'.

C: This line of code modifies the original `lunch_order_df DataFrame`, but the problem statement says that a copy should be made with the new row and the original should be left unchanged.

D: This is the correct answer. The `data Series` is made with the correct information, that is the labels, data entries, and name attribute. Then the `append()` function will make a copy of the calling `DataFrame` with the new row appended.

---


What is the result of the follwing lines of code?

```python
lunch_order_df.drop('price', axis='columns', inplace=True)
```

A: There is a syntax error in this line of code and it will not execute.

B: The column 'price' is dropped from the original `lunch_order_df DataFrame`.

C: A copy of `lunch_order_df` is made without the column 'price' from the original `DataFrame`. The original `DataFrame` is left unchanged.

D: A copy of `lunch_order_df` is made without the row 'price' from the original `DataFrame`. The original `DataFrame` is left unchanged.

**Correct Answer**

B: The column 'price' is dropped from the original `lunch_order_df DataFrame`.

**Explanation**

A: The line of code has no errors.

B: This is the correct answer. The `drop()` function will by default create a copy of the calling `DataFrame`  and drop rows, but the `axis` parameter is set to 'columns' and the `inplace` parameter is set to `True`.

C: The  `inplace` parameter is set to `True`, therefore the original `DataFrame` will be changed.
